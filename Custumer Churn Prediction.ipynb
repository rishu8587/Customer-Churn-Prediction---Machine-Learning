# Initilizing X And Y Values

```python
x = df.drop("Churn", axis = 1)
y = df["Churn"]
```

# Train And Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

# Baseline Model

```python
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
baseline_pree = np.zeros(len(y_test))
baseline_accc = accuracy_score(y_test , baseline_pree)

baseline_accc*100
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_sacled = scaler.fit_transform(x_train)
x_test_sacled = scaler.transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train_sacled,y_train)
```

# Prediction 

```python
y_pred = lr.predict(x_test_sacled)
y_prob = lr.predict_proba(x_test_sacled)
```

# Evalution

```python
from sklearn.metrics import confusion_matrix, classification_report , roc_auc_score , precision_score , recall_score , f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
roc_auc = roc_auc_score(y_test,y_pred)
roc_auc*100
```
